```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np

# List of RGB values
rgb_values = np.array([
    (231,  98,  84), (239, 138,  71), 
    (247, 170,  88), (255, 208, 111),
    (255, 230, 183), (170, 220, 224), 
    (114, 188, 213), ( 82, 143, 173),
    ( 55, 103, 149), ( 30,  70, 110)]) / 255
rgb_values = np.concatenate((rgb_values, np.ones((len(rgb_values), 1))), axis=1)

# Create a custom colormap
custom_cmap = ListedColormap(rgb_values)

# Plot a sample colorbar using the custom colormap
plt.colorbar(plt.cm.ScalarMappable(cmap=custom_cmap))
plt.show()
```
