```python
%load_ext autoreload
%autoreload 2

import glob, os, re, pickle
import sys
import torch

import matplotlib.pyplot as plt
import numpy as np

from inxss import PsiMask
from inxss.psimask import downsample_4d_with_map_coordinates
```

```python
map_path = '/pscratch/sd/z/zhantao/inxs_steering/La2NiO4_bool'
```

```python
psi_mask = PsiMask(map_path, scale_factor=None, device='cpu', preload=False)
```

```python
mask = psi_mask.get_mask(65.4)
```

```python
%timeit psi_mask.get_mask(0.5)
```

```python
print(mask.sum() / mask.numel())
plt.figure()
plt.imshow(mask.sum(axis=-1).sum(axis=-1) > 0)
```

```python
print(mask.sum() / mask.numel())
plt.figure()
plt.imshow(mask.sum(axis=-1).sum(axis=-1) > 0)

downsampled_mask = downsample_4d_with_map_coordinates(mask.float().cpu().numpy(), 0.5, order=0)
print(downsampled_mask.sum() / np.prod(downsampled_mask.shape))
plt.figure()
plt.imshow(downsampled_mask.sum(axis=-1).sum(axis=-1) > 0)
```
