# Utility computation
Compute utility as the function of experiment settings ($\psi$ here)

```python
# output_stack = []
var_calculator = OnlineVariance((np.prod(psi_mask.hkw_grid.shape[:-1]),1), device='cpu')
with torch.no_grad():
    for param in tqdm(particle.positions.T):
        input = psi_mask.get_model_input(param).to(device)
        output = model(input).view(-1, 1)
        var_calculator.update(output.detach().cpu())
        # output_stack.append(output.detach().cpu())

std = var_calculator.stds().view(psi_mask.hkw_grid.shape[:-1])

utility = []
for _angle in tqdm(psi_mask.psi_grid):
    _mask = psi_mask.load_memmap_mask(_angle)
    utility.append((_mask.sum(dim=2) * std).mean())
utility = torch.stack(utility)
```

```python
fig, ax = plt.subplots(1,1)
ax.plot(psi_mask.psi_grid, utility)
ax.set_xlabel('angle')
ax.set_ylabel('utility')
```

```python
utility_diff = (utility - utility.min()) / (utility.max() - utility.min())

import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch

# Create a figure and an axis with a polar projection
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

# Plot data on the polar axis
ax.plot(psi_mask.psi_grid / 360 * 2 * np.pi, utility_diff, color='r', linewidth=2)

arc_arrow(ax, start_angle=55, end_angle=80, radius=utility_diff.max() * 1, arrow_size=0.025, label='$\psi$')

rad_arrow(ax, angle=135/2, start_radius=utility_diff.max() * 0.85, end_radius=utility_diff.max() * 1.15, arrow_size=0.025, label='$U(\psi)$')

ax.set_rlim([0, 1.2])

# Display the figure
plt.show()
```

```python
loss_by_angle = []
for next_angle in tqdm(np.linspace(0, 360, 30)):
    next_mask = psi_mask.load_memmap_mask(next_angle)

    # measurement = calc_Sqw_from_Syy_Szz(psi_mask.hklw_grid[next_mask], experiment.Syy_func, experiment.Szz_func)
    measurement = experiment.get_measurements_by_mask(next_mask, poisson=False)
    # max threshold of the sub-mask that will be compared with neural network predictions
    threshold = 0.05
    threshold_mask = measurement > measurement.max() * threshold

    # output_stack = []
    # var_calculator = OnlineVariance((np.prod(psi_mask.hkw_grid.shape[:-1]),1), device='cpu')
    model.to(device)

    dummy_x_input, dummy_l_input = psi_mask.get_model_input(torch.zeros(2), grid='hklw')
    dummy_x_input = dummy_x_input[next_mask][threshold_mask]
    dummy_l_input = dummy_l_input[next_mask][threshold_mask]

    x_input = dummy_x_input.unsqueeze(0).repeat(particle.num_particles, 1, 1)
    l_input = dummy_l_input.unsqueeze(0).repeat(particle.num_particles, 1, 1)

    with torch.no_grad():
        for i_param, param in enumerate(particle.positions.T):
            x_input[i_param,...,-2:] = param
            # if i_param == 10:
            #     break
    predictions = model(x_input.to(device), l_input.to(device)).detach().cpu()
    # _loss = (experiment.neutron_flux * predictions - measurement[threshold_mask][None]).norm(dim=-1)
    _loss = torch.nn.functional.poisson_nll_loss(
        experiment.neutron_flux * predictions.clip(0.), 
        measurement[threshold_mask][None], log_input=False, reduction="none").mean(dim=-1)
    loss_by_angle.append(_loss)
```
